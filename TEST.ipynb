```python
def load_model(model_file):
    """

    :param model_file:
    :return:
    """
    model = joblib.load(model_file)
    return model


def get_prediction_proba(model, query):
    """

    :param model:
    :param query:
    :return:
    """
    probability = model.predict_proba(query)
    return probability
```
